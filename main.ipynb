```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


df = pd.read_csv('movies.csv')
print(df.head())
```

### Построить любой график (любого рода) с помощью чистого matplotlib / pylab

```python
plt.hist(df['duration'], bins=20, alpha=0.5, color='b')
plt.xlabel('duration')
plt.ylabel('score')
plt.title('Score from duration')
plt.show()
```

### Построить несколько суб-графиков на одном графике используя .subplot.

```python
# построение точечной диаграммы и гистограммы
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))


# histogram
ax2.hist(df['duration'], bins=20, alpha=0.5, color='b')
ax2.set_xlabel('duration')
ax2.set_ylabel('score')
ax2.set_title('Score from duration')

# график рассеяния
ax1.scatter(df['badReviews'],df['goodReviews'], alpha=0.5, color='r')
ax1.set_xlabel('badReviews')
ax1.set_ylabel('goodReviews')
ax1.set_title('Bad reviews and good reviews')

plt.show()
```

### На основе Вашего датасета построить три разных вида графиков используя pandas.DataFrame.plot / Series.plot

```python
df.plot(x="duration", y=["score"])
plt.show()
```

### График разброса (scatter plot) - отображает точки на плоскости, где каждая точка представляет две числовые переменны

```python
df.plot.scatter(x='duration', y='badReviews', figsize=(8, 5))
plt.xlabel('Duration')
plt.ylabel('Bad reviews')
plt.title('Duration and Bad reviews')
plt.show()
```

### Boxplot (ящик с усами) - отображает показатели центральной тенденции, размах и выбросы

```python
# Boxplot для столбца 'continent' и 'beer_servings'
df.boxplot(column='duration', by='color', figsize=(8, 5))
plt.xlabel('Color')
plt.ylabel('Duration')
plt.title('Duration by Color')
plt.suptitle('')  # Удаление заголовка, сгенерированного методом boxplot
plt.show()
```

### Построить .pairplot с помощью seaborn на каком-либо наборе данных, проанализировав взаимосвязь набора числовых переменных

```python
sns.pairplot(data=df, vars=['duration', 'badReviews', 'neutralReviews', 'goodReviews'])
```

### Построить .jointplot, проанализировав взаимосвязь двух конкретных числовых переменных

```python
sns.jointplot(data=df, x='duration', y='score', kind='reg')
```

### Построить .boxplot или .violinplot на каком-либо наборе данных, проанализировав какой-то числовой показатель в разрезе категориального

```python
sns.boxplot(data=df, x='score', y='country')
```

### Построить график heatmap попарной корреляции всех числовых признаков (категориальные признаки убираете из графика)

```python
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
```

### Построить любой график (связанный с вашими данными) с помощью plotly
